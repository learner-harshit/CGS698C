```
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pymc3 as pm
import scipy
from sklearn.preprocessing import StandardScaler
```

```
'''
A :
Creating a sample space of 4096 tosses .
Let us assume that the coin is bias with p = 0.38 where p is the probability of getting a head
'''

outcomes = np.random.binomial( n = 1, p = 0.38, size = 4096 )
```

```
# Function to plot beta prior 

def plot_prior(alpha, beta, ax=None):
    x = np.linspace(0, 1, 1000)
    y = scipy.stats.beta.pdf(x, alpha, beta)

    if not ax:
        fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel(r"$\theta$", fontsize=20)
    ax.set_ylabel(r"$P(\theta)$", fontsize=20)
    ax.set_title("Prior: Beta({},{})".format(alpha,beta));
```

```
#Funtion to plot posterior

def plot_posterior(heads, tails, alpha, beta, ax=None):
    x = np.linspace(0, 1, 1000)
    a = heads + alpha
    b = tails + beta
    y = scipy.stats.beta.pdf(x, a, b)
        
    if not ax:
        fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel(r"$\theta$", fontsize=20)
    ax.set_ylabel(r"$P(\theta|D)$", fontsize=20)
    ax.set_title("Posterior after {} draws : {} heads, {} tails, \
                 Prior: BetaPDF({},{})".format(heads+tails, heads, tails, alpha, beta));

```

```

#Part A : Uninformed Prior
a = 1
b = 1
plot_prior(a,b)
for n in [1,2,3,4,100,1024,4096]:
  outcome = outcomes[:n-1]
  heads = np.count_nonzero(outcome)
  tails = n - heads
  plot_posterior(heads, tails, a, b )
```

```
#Part B : Prior with peak at 0.25
a = 4
b = 10
plot_prior(a,b)
for n in [1,2,3,4,100,1024,4096]:
  outcome = outcomes[:n-1]
  heads = np.count_nonzero(outcome)
  tails = n - heads
  plot_posterior(heads, tails, a, b )
```
